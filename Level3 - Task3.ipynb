# Step 1:

# Step 2: 
   Distribution of Ratings using Histogram

Explanation: We import the libraries and load the cleaned dataset to start exploring.

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Aggregate rating'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

# Step 3:

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Aggregate rating', data=df, palette='mako')
plt.title('Count of Each Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.show()
```

Explanation: This bar plot shows how many restaurants received each rating value.

4.2 Create bar plot of top cuisines by average rating

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv(r"C:\Users\HP\Downloads\Dataset_cleaned .csv")

# Handle multi-cuisine entries
df_cuisine = df.copy()
df_cuisine['Cuisines'] = df_cuisine['Cuisines'].astype(str)
df_cuisine = df_cuisine.assign(Cuisines=df_cuisine['Cuisines'].str.split(', ')).explode('Cuisines')

# Calculate average rating per cuisine
avg_rating_cuisine = df_cuisine.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=avg_rating_cuisine.values, y=avg_rating_cuisine.index, palette='crest')
plt.title('Top 10 Cuisines by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Cuisine')
plt.show()
```

Explanation: We display the top 10 cuisines based on their average rating.

# Step 5: 
   Compare Average Ratings by City

```python
avg_rating_city = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=avg_rating_city.values, y=avg_rating_city.index, palette='flare')
plt.title('Top 10 Cities by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('City')
plt.show()
```

Explanation: This bar plot helps compare which cities have the highest-rated restaurants.

# Step 6: 
   Relationship - Votes vs Ratings

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Aggregate rating', y='Votes', data=df, alpha=0.6)
plt.title('Votes vs Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Votes')
plt.show()
```

Explanation: This scatter plot checks if higher-rated restaurants receive more customer votes.

# Step 7: 
   Relationship - Price Range vs Rating

```python
plt.figure(figsize=(7,5))
sns.boxplot(x='Price range', y='Aggregate rating', data=df, palette='coolwarm')
plt.title('Rating Distribution by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Aggregate Rating')
plt.show()
```

Explanation: This boxplot helps understand how ratings vary across different price ranges.
